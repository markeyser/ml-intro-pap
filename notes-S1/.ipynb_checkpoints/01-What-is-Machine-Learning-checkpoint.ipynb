# What is Machine Learning?

## Introduction

Machine learning is a branch of artificial intelligence that deals with learning patterns and rules from training data. Its origins date back to the middle of the last century but it's really in the last decade that companies have used it massively for their products.

This revolution of machine learning has been enabled by really three factors:

- First of all memory storage has become really cheap and accessible.

- Second computing power has also become really cheap and accessible.

- And third, sensors, phones and web application have produced a lot of data which has contributed to training these machine learning models.

## Machine Learning Problems

There are different types of machine learning including:

- Supervise Learning
- Unsupervise Lerning
- Reinforcement Learning 

On this course will primarily focus on supervised learning. But it is important to understand what each of these is. So let's start with supervised learning 

## Supervised Machine Lerning

In supervised learning an algorithm learns from label data.

For example let's say we're training an image recognition algorithm to distinguish cats from dogs. 

![](../imgs/cat-and-dog.png)

Each training data point will be a pair of an image (that's the training data) and the label, specifying whether it's a cat or a dog.

![](../imgs/classfication-1.png)

In a similar way if we're training a translation engine we will provide both input and output sentences
and ask the algorithm to learn the function that connects them.

## Usupervised Learning

Converselly, in unsupervised learning data comes without label, and the task is to find similar data points in the data set in order to identify any underlying hidden structure.

For example think of a dataset that contains purchase preferences from e-commerce users.These users will likely form clusters  that behave in a similar way and they may have seen similar purchase behavior.

![](../imgs/unsupervised.png)

They may spend the same amount of money or a similar amount of money or buy similar type of objects.

You can think of these different groups as different tribes different preferences.

![](../imgs/unsupervised2.png)

Once these tribes are identified we can describe each data point, which is each user, in terms of the
tribe it belongs to gaining a deeper understanding of the data and also reducing the number of data
points that we have to do with.

## Reinforced Learning

Finally reinforcement learning is similar to supervised learning. But in this case the algorithm is training an agent to act in an environment. The agent can take some action and know the state of the environment and these actions lead to an outcome. The outcome is attached to a score, and the algorithm tries to maximize the score.

Typical examples are the algorithms that learned to play games like chess or go. 

![](../imgs/go.png)

And the main difference with supervised learning is that the algorithm does not really see a label for each action but may receive the score after a sequence of actions.

In 2016, a software trained with this technique beat the world champion and marked the new milestone in the race towards artificial intelligence.


So in conclusion we have seen there are many types of machine learning problems and deep learning has been successfully applied to both supervised unsupervised and reinforcement learning problems. We will focus on supervised learning problems in the rest of the course.

## Supervise Learnig Successful Applications

Let's see some successful applications of supervised learning.

I'd like to start with an example. Have you ever noticed that e-mail spam is practically non-existent.

![](../imgs/spam.png)

This is one of the great successes of supervised learning. 

In the early 2000s, E-mail boxes were plagued by ton of e-mails advertising pills and utter fraudulent activity moneymaking schemes and in general very useless information.

![](../imgs/spam3.png)

The first step to get rid of these was to allow users to move spam emails into a spam folder.

![](../imgs/spam2.png)

This action provided the training labels. 

Then, with millions of users manually cataloging spam large email providers like Google and Yahoo could quickly gather enough examples of what spam looks like to train a model that would predict the probability for a message to be spam.

![](../imgs/spam4.png)

That's an example of what's called a binary classifier and this is a machine learning algorithm that
learns to distinguish between two classes like: 

- True or false 
- Dead or alive or 
- negative or positive 

## Types of supervised Lerning

- Binary classifiers
- Multi-class classifiers
- Continuous (Regression)

### Binary Classifiers

Binary classifiers trained with supervised learning are present everywhere. You find them in predicting:

- Churn prediction
- Sentiment analysis
- Click predition
- Desease screening
- many many more ...

#### Churn example

Churn for example it is a very common application in telecom. When you want to predict if the user is about to go to a competitor or not it's a yes no type of question. And this model is useful to know when and to whom to make an offer for a new product.

#### Sentiment analysis example 

Another application is sentiment analysis. This is what a social media company would do if you are for example a celebrity that has millions of fans in its Facebook page or in social media page. You post something and you will receive a million of comments and you want to quickly know the average sentiment about your post was.

And the question is how can you know if your followers were primarily happy or angry or perplexed about
what you tweeted. A sentiment classifier can tell you that for each single comment and therefore
you can get an overall reaction by aggregating the responses.

#### Other aplications of binary classifiers

Other applications of binary classifiers include predicting whether a user is going to click or not
on a link, screening the presence of disease, detecting if you are a human or a bot. Think of captchas.

### Muti-class classifiers

A natural extension of binary classifier is the case we're more than

Supervised learning is also used to predict continuous quantities:
