# Outliers

In this lesson, I will give you an introduction about what outliers are, which machine learning models they affect, and how we can identify outliers in a variable.

## Outliers definition

- An outlier is a data point which is significantly different from the remaining data.
- "An outlier is an observation which deviates so much from the other observations as to arouse suspicions that it was generated by a different mechanism." [D. Hawkins. Identification of Outliers, Chapman and Hall, 1980.]

## Should outliers be removed?

So, outliers are values that are so different to th remaining data for that variable. Should we remove them from the dataset? 

![](../imgs/outliers.png)

Well, depending on the context, outliers either deserve special attention or should be completely ignored.

Let's see, for example, in revenue forecasting. If an unusual spike of revenue is observed, it is probably a good idea to pay extra attention and try and figured out what caused that spike. 

In the same way, if an unusual transaction in the credit card appears we may want a different look at it because maybe a sign of fraudulent activity. 

So, in instances like these, it is useful to look for and investigate further those outlier values.

If outliers, however, are introduced by mechanical error or measurement error or anything else that cannot be generalizable it is a good idea to filter out those outliers before feeding the data into a modeling algorithm.
